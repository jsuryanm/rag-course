### Creating the document objects

```python
from langchain_core.documents import Document 

doc = Document(
    page_content="This is the actual content of the document. It contains the text we want to process.",
    metadata={
        "source": "example.pdf",
        "page": 1,
        "author": "John Doe",
        "date": "2025-01-15"
    }
)

# Inspect the document
print("📄 Document Structure:")
print(f"\nType: {type(doc)}")
print(f"\nContent (first 100 chars): {doc.page_content[:100]}...")
print(f"\nMetadata: {doc.metadata}")
print(f"\nSource: {doc.metadata['source']}")
print(f"Page Number: {doc.metadata['page']}")
```

### Loading PDF files

```python
from langchain_community.document_loaders import PyPDFLoader
from pathlib import Path

pdf_path = "attention.pdf"

if Path(pdf_path).exists():
    print("Loading pdf")

    loader = PyPDFLoader(pdf_path)
    docs = loader.load()
    print(f"Loaded {len(docs)} pages")

    print("First page:")
    print(f"Metadata: {docs[0].metadata}")
    print(f"\n📄 Last Page (page {len(docs)}):")
    print(f"   Content (first 200 chars): {docs[-1].page_content[:200]}...")
    
else:
    print(f"❌ PDF not found: {pdf_path}")
    print("   Make sure the file exists in the project root")
```

### Lazy loading for large pdfs 


```python
if Path(pdf_path).exists():
    loader = PyPDFLoader(pdf_path)

    print("Lazy loading pages")

    for i,page in enumerate(loader.lazy_load()):
        if i >= 5:
            break 
        print(f"\nPage {i + 1}")
        print(f"Length: {len(page.page_content)} characters")
        print(f"Preview: {page.page_content}")
    print("\n💡 Tip: Use lazy_load() for PDFs > 100 pages to save memory")
```

### Loading multiple PDF's from a Directory

```python
pdf_directory = "pdfs"

if Path(pdf_directory).exists():
    print(f"Loading documents from {pdf_directory}\n")

    all_docs = []

    pdf_files = list(Path(pdf_directory).glob("*.pdf"))
    print (f"Found {len(pdf_files)} PDF files")

    for pdf_file in pdf_files:
        print(f" - {pdf_file.name}")

        loader = PyPDFLoader(str(pdf_file))

        docs = loader.load()
        all_docs.extend(docs)
        print(f"Loaded {len(docs)} pages from PDF's")

    sources = set(doc.metadata['source'] for doc in all_docs)
    print("\nSources:")
    for source in sources:
        print(f"- {Path(source).name}")

else:
    print(f"Directory not found: {pdf_directory}")
```

### Loading CSV Files

**CSVLoader** converts each row of csv file into seperate Document

```python
from langchain_community.document_loaders import CSVLoader

csv_path = "sample_data/products.csv"

if Path(csv_path).exists():
    print(f"Loading CSV: {csv_path}")

    loader =  CSVLoader(csv_path,
                        source_column="product_name")
    
    docs = loader.load()

    print(f"Loaded {len(docs)} products\n")

    for i,doc in enumerate(docs[:3],1): # starts counting from 1, instead of 0
        print(f"{'='*70}")
        print(f"Product {i}:")
        print(f"{'='*70}")
        print(doc.page_content)
        print(f"\nSource:{doc.metadata['source']}")
        print(f"Row:\n{doc.metadata.get("row","N/A")}")
        print()
    print(f"... and {len(docs) - 3} more products")
else:
    print(f"CSV not found: {csv_path}")
```

### Loading JSON files 

```python
from langchain_community.document_loaders import JSONLoader

json_path = "sample_data/api_response.json"

if Path(json_path).exists():
    print(f"Loading JSON: {json_path}\n")

    loader = JSONLoader(file_path=json_path,
                        jq_schema=".articles[]",
                        text_content=False) # returns full json for each article
    
    documents = loader.load()

    print(f"Loaded {len(documents)} articles\n")

    print("First article:")
    print(f"Content: \n {documents[0].page_content}\n")
    print(f"Metadata: {documents[0].metadata}")

else:
    print(f"JSON not found")
```

### Extracting Specific Fields from JSON

```python
if Path(json_path).exists():
    loader = JSONLoader(file_path=json_path,
                        jq_schema=".articles[].content", 
                        # goes inside the JSON and extract "article":[{"title":"A"}] part
                        # this will extract only content part 
                        text_content=True)
    
    docs = loader.load()

    for i,doc in enumerate(docs[:2],1):
        print(f"{i}, {doc.page_content[:150]}")
        print()
```

```python
from langchain_community.document_loaders import UnstructuredHTMLLoader

html_path = "sample_data/blog_post.html"

if Path(html_path).exists():
    print(f"Loading HTML: {html_path}\n")

    loader = UnstructuredHTMLLoader(html_path)

    docs = loader.load()
    print(f"Loaded {len(docs)} documents")

    doc = docs[0]
    print(f"Content length: {len(doc.page_content)} chars")
    print(f"\n📝 First 500 characters:\n{doc.page_content[:500]}...")
    print(f"\n🔍 Metadata: {doc.metadata}")
    
else:
    print(f"❌ HTML not found: {html_path}")
```

### Loading multiple urls

```python
from langchain_community.document_loaders import WebBaseLoader

urls = [
    "https://python.langchain.com/docs/introduction/",
    "https://python.langchain.com/docs/expression_language/"
]

loader = WebBaseLoader(urls)
docs = loader.load()

print(f"Loaded {len(docs)} pages")
for doc in docs:
    print(f" - {doc.metadata['source']}")

```

### Text files

```python
from langchain_community.document_loaders import TextLoader 

txt_path = "sample_data/notes.txt"

if Path(txt_path).exists():
    loader = TextLoader(txt_path)
    docs = loader.load()

    print(f"Loaded {len(docs)} documents")
    doc = docs[0]
    print(f"Content length: {len(doc.page_content)} chars")
    print(f"First 300 chars:\n{doc.page_content[:300]}")
    print(f"\nmetadata:{doc.metadata}")
else:
    print("Text file not found")
```

### Batch loading with `DirectoryLoader`

```python
from langchain_community.document_loaders import DirectoryLoader

data_dir = "sample_data"

if Path(data_dir).exists():
    print(f"Loading all text files from {data_dir}")

    loader = DirectoryLoader(data_dir,
                             glob="*.txt",
                             loader_cls=TextLoader,
                             show_progress=True)
    
    docs = loader.load()
    print(f"Loaded {len(docs)} text files")

    for doc in docs:
        print(f" - {doc.metadata['source']} {len(doc.page_content)} chars")

else:
    print(f"Directory not found: {data_dir}")
```

When to use which 

- Academic papers? → `PyPDFLoader`
- 📊 Structured data? → `CSVLoader`
- 🔧 API data? → `JSONLoader`
- 🌐 Web content? → `WebBaseLoader`
- 📝 Simple text? → `TextLoader`
- 📂 Entire folder? → `DirectoryLoader`
